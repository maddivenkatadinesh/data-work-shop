```python
x_axis =[2,5,8,15]
y_axis =[4,9,16,22]
plt.plot(x_axis,y_axis)
# 
```

```python
x_axis =[2,5,8,15]
y_axis =[4,9,16,22]
plt.plot(x_axis,y_axis,label='Line Plot')
plt.title("Line plotting")
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.legend(loc = "best")
```
