## Robust Scalar
* if your dataset contains more no.of.outliers, then robust scalar is best for scaling

```python
#        xi  -  q2(X)
#     -----------------
#      q3(x) -  q1(x)

### Inter Quartile Range(IQR)
# q1--->25%(Lower Quartile)
# q2--->50%(Medium Quartile)
# q3--->75%(Upper Quartile)
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
rs
```

```python
rs2 = rs.fit_transform(data1)
rs2
```

```python
rs2= pd.DataFrame(rs2, columns=data1.columns)
rs2.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))
sns.kdeplot(data1['TV'], ax=ax1)
sns.kdeplot(data1['radio'], ax=ax1)
sns.kdeplot(data1['newspaper'], ax=ax1)
sns.kdeplot(data1['sales'], ax=ax1)

sns.kdeplot(rs2['TV'], ax=ax2)
sns.kdeplot(rs2['radio'], ax=ax2)
sns.kdeplot(rs2['newspaper'], ax=ax2)
sns.kdeplot(rs2['sales'], ax=ax2)
```

## standard Scaler

* Most commonly used scalling technique. it scales between 0 to mean

```python
# standard scalar formula


#    xi - mean(x)
#  -----------------
#      std(x)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss2 = ss.fit_transform(data1)
ss2 = pd.DataFrame(ss2,columns=data1.columns)
ss2.head()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))
sns.kdeplot(data1['TV'], ax=ax1)
sns.kdeplot(data1['radio'], ax=ax1)
sns.kdeplot(data1['newspaper'], ax=ax1)
sns.kdeplot(data1['sales'], ax=ax1)

sns.kdeplot(ss2['TV'], ax=ax2)
sns.kdeplot(ss2['radio'], ax=ax2)
sns.kdeplot(ss2['newspaper'], ax=ax2)
sns.kdeplot(ss2['sales'], ax=ax2)
```
